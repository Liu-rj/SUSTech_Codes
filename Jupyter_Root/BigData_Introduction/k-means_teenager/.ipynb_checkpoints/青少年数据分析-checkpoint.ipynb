```python
teenager["gender"]=teenager["gender"].replace(np.nan,"unknown")
```

```python
teenager["gender"].value_counts()
```

```python
gender_dummies=pd.get_dummies(teenager["gender"],prefix="gender")
gender_dummies.head(10)
```

```python
teenager=pd.concat([teenager,gender_dummies],axis=1)
```

```python
teenager.head(10)
```

```python
age_mean=teenager["age"].mean()
```

```python
teenager["age_avg_imputated"]=teenager["age"].fillna(value=age_mean)
# teenager["age"][np.isnan(teenager["age"])]=age_mean
```

```python
teenager.head(10)
```

```python
teenager_filtered = teenager.drop(['gender','age'],axis=1)
```

```python
from sklearn import preprocessing
teenager_zscore=pd.DataFrame(preprocessing.scale(teenager_filtered), columns=teenager_filtered.columns)
teenager_zscore.head(5)
```

```python
from sklearn.cluster import KMeans
teenager_cluster_model = KMeans(n_clusters = 5)
teenager_cluster_model.fit(teenager_zscore)
```

```python
teenager_labels = pd.Series(teenager_cluster_model.labels_)
teenager_labels.value_counts().sort_index()
```

```python
teenager_centers = pd.DataFrame(teenager_cluster_model.cluster_centers_,columns=teenager_zscore.columns)
teenager_centers
```

```python
centers_t = teenager_centers.T # transpose matrix
centers_t.columns = ["cluster_0", "cluster_1", "cluster_2", "cluster_3", "cluster_4"]
centers_t["cluster_0"].sort(ascending = False, inplace=False).head(10)
```
