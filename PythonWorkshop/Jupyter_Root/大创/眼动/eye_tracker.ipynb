## 使用原始数据gaze point进行离群值剔除、聚类、得到中心点

```python
sns.set()
plt.figure(figsize=(4,4), dpi=144)
sns.scatterplot(x='Gaze point X', y='Gaze point Y', data=gaze_valid[['Gaze point X', 'Gaze point Y']])
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
```

# 眼动数据处理

```python
def extract(temp):
    temp = temp[['Gaze point X [MCS px]', 'Gaze point Y [MCS px]', 'Pupil diameter left [mm]', 'Pupil diameter right [mm]']].dropna().reset_index(drop=True)
    xcenter = temp['Gaze point X [MCS px]'].sum() / 2475
    ycenter = temp['Gaze point Y [MCS px]'].sum() / 2475
    temp['distance'] = np.sqrt(np.square(temp['Gaze point X [MCS px]'] - xcenter) + np.square(temp['Gaze point Y [MCS px]'] - ycenter))
    temp = temp[['Pupil diameter left [mm]', 'Pupil diameter right [mm]', 'distance']]
    return temp

def dump(data, offset, onlysee, seeImagine, onlyImagine, relax):
    for i in range(10):
        temp = data[(data['Recording timestamp [ms]'] >= offset) & (data['Recording timestamp [ms]'] < (offset + 5))]
        onlysee = pd.concat([onlysee, extract(temp)], axis=0)
        temp = data[(data['Recording timestamp [ms]'] >= (offset + 7.5)) & (data['Recording timestamp [ms]'] < (offset + 12.5))]
        relax = pd.concat([relax,extract(temp)],axis=0)

        offset = offset + 15
        temp = data[(data['Recording timestamp [ms]'] >= offset) & (data['Recording timestamp [ms]'] < (offset + 5))]
        seeImagine = pd.concat([seeImagine, extract(temp)], axis=0)
        temp = data[(data['Recording timestamp [ms]'] >= (offset + 7.5)) & (data['Recording timestamp [ms]'] < (offset + 12.5))]
        relax = pd.concat([relax,extract(temp)],axis=0)
        offset = offset + 15

        temp = data[(data['Recording timestamp [ms]'] >= offset) & (data['Recording timestamp [ms]'] < (offset + 5))]
        onlyImagine = pd.concat([onlyImagine, extract(temp)], axis=0)
        temp = data[(data['Recording timestamp [ms]'] >= (offset + 7.5)) & (data['Recording timestamp [ms]'] < (offset + 12.5))]
        relax = pd.concat([relax,extract(temp)],axis=0)
        offset = offset + 15
    return onlysee, seeImagine, onlyImagine, relax
        
def import_data(string, offset):
    data = pd.read_excel(string)
    data['Recording timestamp [ms]'] =(data['Recording timestamp [ms]'] / 1000)
    data = data[data['Recording timestamp [ms]'] >= offset]
    return dump(data, offset, onlysee, seeImagine, onlyImagine, relax)
```

```python
def dump_5(data, offset, onlysee, seeImagine, onlyImagine, relax):
    for i in range(5):
        temp = data[(data['Recording timestamp [ms]'] >= offset) & (data['Recording timestamp [ms]'] < (offset + 5))]
        onlysee = pd.concat([onlysee, extract(temp)], axis=0)
        temp = data[(data['Recording timestamp [ms]'] >= (offset + 7.5)) & (data['Recording timestamp [ms]'] < (offset + 12.5))]
        relax = pd.concat([relax,extract(temp)],axis=0)

        offset = offset + 15
        temp = data[(data['Recording timestamp [ms]'] >= offset) & (data['Recording timestamp [ms]'] < (offset + 5))]
        seeImagine = pd.concat([seeImagine, extract(temp)], axis=0)
        temp = data[(data['Recording timestamp [ms]'] >= (offset + 7.5)) & (data['Recording timestamp [ms]'] < (offset + 12.5))]
        relax = pd.concat([relax,extract(temp)],axis=0)
        offset = offset + 15

        temp = data[(data['Recording timestamp [ms]'] >= offset) & (data['Recording timestamp [ms]'] < (offset + 5))]
        onlyImagine = pd.concat([onlyImagine, extract(temp)], axis=0)
        temp = data[(data['Recording timestamp [ms]'] >= (offset + 7.5)) & (data['Recording timestamp [ms]'] < (offset + 12.5))]
        relax = pd.concat([relax,extract(temp)],axis=0)
        offset = offset + 15
    return onlysee, seeImagine, onlyImagine, relax
        
def import_data_5(string, offset):
    data = pd.read_excel(string)
    data['Recording timestamp [ms]'] =(data['Recording timestamp [ms]'] / 1000)
    data = data[data['Recording timestamp [ms]'] >= offset]
    return dump_5(data, offset, onlysee, seeImagine, onlyImagine, relax)
```

# 距离+瞳孔

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

x=temp[['distance', 'Pupil diameter left [mm]', 'Pupil diameter right [mm]']]
y=temp['target']

scaler = StandardScaler()
scaler.fit(x)
x= scaler.transform(x)

x_train,x_test,y_train,y_test=train_test_split(x,y)

from sklearn.svm import SVC
from sklearn import metrics

kernelList = ['linear','rbf','sigmoid']         

for kernel in kernelList:
    svc = SVC(kernel=kernel).fit(x_train,y_train)
    y_pred = svc.predict(x_test)
    score_svc = metrics.accuracy_score(y_test,y_pred)
    print(score_svc)
```

```python
score_gamma=[]
gammaList=np.logspace(-10,1,50)

# for gamma in gammaList:
#     svc = SVC(gamma=gamma).fit(x_train,y_train)
#     score_gamma.append(svc.score(x_test,y_test))

svc = SVC(gamma=100).fit(x_train,y_train)
score_gamma.append(svc.score(x_test,y_test))

print('gamma={}时,得分最高={}'.format(gammaList[score_gamma.index(max(score_gamma))],max(score_gamma)))

# plt.plot(gammaList,score_gamma)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

lr = LogisticRegression()                                        # 实例化一个LR模型
lr.fit(x_train,y_train)                                          # 训练模型
y_prob = lr.predict_proba(x_test)[:,1]                           # 预测1类的概率
y_pred = lr.predict(x_test)                                      # 模型对测试集的预测结果
fpr_lr,tpr_lr,threshold_lr = metrics.roc_curve(y_test,y_prob)    # 获取真阳率、伪阳率、阈值
auc_lr = metrics.auc(fpr_lr,tpr_lr)                              # AUC得分
score_lr = metrics.accuracy_score(y_test,y_pred)                 # 模型准确率
print([score_lr,auc_lr])
```

```python
from sklearn import tree
from sklearn import metrics

x=temp[['distance', 'Pupil diameter left [mm]', 'Pupil diameter right [mm]']]
y=temp['target']
x_train,x_test,y_train,y_test=train_test_split(x,y)

dtc = tree.DecisionTreeClassifier()                              # 建立决策树模型
dtc.fit(x_train,y_train)                                         # 训练模型
y_prob = dtc.predict_proba(x_test)[:,1]                          # 预测1类的概率
y_pred = dtc.predict(x_test)                                     # 模型对测试集的预测结果 
fpr_dtc,tpr_dtc,threshod_dtc= metrics.roc_curve(y_test,y_prob)   # 获取真阳率、伪阳率、阈值
score_dtc = metrics.accuracy_score(y_test,y_pred)                
auc_dtc = metrics.auc(fpr_dtc,tpr_dtc) 
print([score_dtc,auc_dtc])
```

```python
depthList=np.arange(1,50)
score=[]

for i in depthList:
    dtc = tree.DecisionTreeClassifier(max_depth=i).fit(x_train,y_train)
    score.append(dtc.score(x_test,y_test))

print('最优深度为{}，最佳得分是{}'.format(depthList[score.index(max(score))],max(score)))
```
